# 39. Combination Sum

[leetcode](https://leetcode.com/problems/combination-sum/)

Given an array of distinct integers candidates and a target integer target, return a list of all unique combinations of candidates where the chosen numbers sum to target. You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the 
frequency
 of at least one of the chosen numbers is different.

The test cases are generated such that the number of unique combinations that sum up to target is less than 150 combinations for the given input.

# Reasoning 

[neetcodevideo](https://www.youtube.com/watch?v=GBKI9VSKdGg)

__Note__: we want combinations here, not _permutations_. So [3,2,2] and [2,2,3] would not be valid. 

So each combination must have a unique number of each numbers.  

A brute force is to build a `decision tree`. However, a classical search binary decision tree where at every branch you chose what to, we get duplicates. So we use a special `recursion tree` and a `depth first search algorithm`

Consider a decision tree, where at a left and right branches we fix that if we use a number (or a number of a given number) in a left subtree, we do not consider the 2 in the second subtree. 

For example 
root -> [with 2; without 2] -> [with 2,2; with 2; without 2; without 2] ... 
Thus, at each branch we allow to use a given number certain number of times only once. 

The lagorithm is recursive and allows to avoid duplicates. 

__NOTE__ we are allowed to add Nothing at a given split. 

We will use a pointer to track which elements from the original tree have been used before. 

The time complexity O(NumberOfSplits ^ HightOfDecisionTree), where Number = 2 and Height = target, so it is O(2^H)




```python
from typing import List
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        res= []
        def dfs(i,cur,total):
            """ 
            i - index of the currently used number (pointer to track what was used) 
            cur - current variable (list of currently added variables)
            total - sum of elements in the cur
            """
            # check if we reached the result
            if total == target:
                # append copy as we will use this combination later
                res.append(cur.copy()) 
                return
            # check if we obershoot
            if i > len(candidates) - 1 or total > target:
                return
            # chose what decision to make

            # LEFT: take i-th from candidates and add to the curr
            cur.append(candidates[i])
            # call dfs on the result. Note we did not change i here
            dfs(i, cur, total + candidates[i])

            # CLEAN: Note, now we need to clean the left path
            cur.pop()

            # RIGHT:, that we skip the i-th now, and do not update cur or total
            dfs(i+1, cur, total)


        dfs(0,[],0)
        return res
            
            
```
