# 104. Maximum Depth of Binary Tree

[leetcode](https://leetcode.com/problems/maximum-depth-of-binary-tree/)

Given the root of a binary tree, return its maximum depth.

A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

# Reasoning

[neetcodevideo](https://www.youtube.com/watch?v=hTM3phVI6YQ&ab_channel=NeetCode)

__NOTE__: this is an example of a _tree traversal problem_ that has many solutions. 

- `Recursive DFS` (simplest solution)
- `Iterative DFS` requries `call stack`
- `BFS` requires Deque `queue` data structure

For a single node -- 0
For a node with one leaf -- 1
for a Given example it is  
$$
1 + \max(DFS(left), DFS(right))
$$
where 1 is the original 

For each node we run DFS and find maximum for left and wright branches. 

This alorithm has time and memory complexities of O(n) as we traversing the tree only once. 

__NOTE__: BFS may be not more efficient in this problem, but it allows to avoid recursion.  
BFS is essencialy a level traversal, where each level is visited by level. It is an _intoitive_ way, as it counts the number of levels. 

__NOTE__: implementeing a BFS algorithm requires a `queue` or a `deck`. HEre a queue consistis of "levels", for each level. The 'queue' is populated as follows, 
- start at a node, add its value to the array (append)
- look at anode and if it has children, replace the appended value with new values for their childred (node you may need to append extra value as you may have 2 childred while previously you appended 1 value). 
- to each appending attach a value representing the depth of the current level

For non-recursive DFS we need a `stack` to implement a `pre-order` DFS. The algorithm is the following:
- start with the node, add children to the stack
    - In the normal _pre-order traversal_ we add the left subtree first.
    - Look at the children of each subtree. If not processed go back to the last 
When traversing the tree this way we can add childred on the stack as well as depth

At each iteration, we add the node the node, and then pop it and replace it with children.  
Visit every node and find the node with largest depth

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
from typing import Optional
from typing import Deque
class Solution:
    def maxDepthRecursive(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        return 1 + max(self.maxDepthRecursive(root.left),
                       self.maxDepthRecursive(root.right))
    
    def maxDepthBFS(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        level = 0
        q = Deque([root])
        while q:
            # go throgh the que and remove every level there

            # take a snapshot of the length of the queue
            # traverse the entire level and add next level
            for i in range(len(q)):
                node = q.popleft()
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
                # each time the qeue is not empty we increment 
                # the level
                level += 1
        return level
    def maxDepthDFSnoRecursion(self, root: Optional[TreeNode]):
        if not root:
            return 0
        stack = [[
            root, # node
            1 # depth
        ]]
        res = 1 # at least one node is there
        while stack:
            node, depth = node.pop()
            # if node is not null 
            if node:
                res = max(res, depth)
                # add children to the stack
                stack.append([node.left,depth+1])
                stack.append([node.right,depth+1])
        return res

    
```
