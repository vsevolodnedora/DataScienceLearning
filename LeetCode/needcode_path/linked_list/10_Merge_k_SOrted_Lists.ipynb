# 23. Merge k Sorted Lists

[neetcode](https://leetcode.com/problems/merge-k-sorted-lists/description/)

You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.

# Reasoning

[neetcodevideo](https://www.youtube.com/watch?v=q5a5OiGbT6Q&ab_channel=NeetCode)

In leetcode 21 we did merging of 2 sorted linked lists. There we created a 'Dummy' node and looped over both linked lists at the same time and appended a new node with the smallest value untill the end.  

An _inefficient_ idea is to merge linked lists progressively. Start with one pair and keep adding new node based on its value. This requires looping through the newly created list each time we want to add a. This would be O(k*n) time comlexity. 

A _more efficient way_ is as in an algorithm `merge sort`, where we merge pairs only, untill all parts used. Then we mrge the pairs among the result from first pairs. This requries al algorithm _merge two sorted linked lists_ to be used.  
In this algorithm, when we go through all nodes at the beginnign, this is O(n) operation. However, we do not need to do it each time when we merge, so each time, after each merger we decrease the list size by 2. Thus, we would need to merge lists only 'log(k)' times. This is the efficnet part. The overall time complexity is O(n*log(k))

The basis of the solution is implementing a function to merge two linked lists

```python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    # def mergeTwoLists(self, l1, l2):
    #     dummy = ListNode(0)
    #     while l1 or l2:
    #         v1 = l1.val if l1 else float('infinity')
    #         v2 = l2.val if l2 else float('infinity')
    #         if v1 < v2:
    #             dummy.next = ListNode(v1)
    #             l1 = l1.next
    #         else:
    #             dummy.next = ListNode(v2)
    #             l2 = l2.next
    #     return dummy.next
    def mergeTwoLists(self, l1, l2):
        dummy = ListNode(0)
        tail = dummy
        while l1 and l2:
            v1 = l1.val
            v2 = l2.val
            if v1 < v2:
                dummy.next = ListNode(v1)
                l1 = l1.next
            else:
                dummy.next = ListNode(v2)
                l2 = l2.next
            dummy = dummy.next 
        if l1:
            dummy.next = l1
        elif l2:
            dummy.next = l2
        return tail.next
    
    def mergeKLists(self, lists: list[Optional[ListNode]]) -> Optional[ListNode]:
        # consider edge cases
        if not lists or len(lists) == 0:
            return None
        # consider pairs of linked lists and merge them
        while len(lists) > 1:
            merged = []
            # select pairs
            for i in range(0, len(lists), 2):
                l1 = lists[i]
                l2 = lists[i+1] if i+1 < len(lists) else None
                merged.append( self.mergeTwoLists(l1,l2) )
            lists = merged
        return lists[0]


    
```
