# 4. Median of Two Sorted Arrays

Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of the two sorted arrays.

The overall run time complexity should be O(log (m+n)).

# Reasoning

[neetcodevideo](https://www.youtube.com/watch?v=q6IEA26hvXc)

The complexity of the problem is that it must be in log(m+n) time.  
This means we cannot just go through all elements one at a time. This would be O(m*n) time.  
This, we _cannot merge_ arrays and get the solution this way.  

Since log time is required, we need _Bibary Search_.  

An important point is the a __median__ of an array is a point _in the middle_ of the array, in terms of its size.  

So in array [1,2,3,4], the median is 3+4/2. In array [1,2,3,4,5] it is just 3, as the array has an odd number of elements. 

The `trick` in this problem is consider _partitions_.  We only need to find the _left partitton_ of two sorted arrays that would include half of the total array. If we find its end, together with the next element (the minimum of two from two partitions), it will be the median.

A partifion is constructred as following. 
- Consider smaller/second array and set `l`, `r` pointers and compute `m` pointer 
- Consider _both_ and devide their size/2 -> int. This gives the _left partition_ of the array. Start with the smaller array. This says how much we need from the larger array to assemble the median.  
- The way how we check if we made the partition right, is the the element on the left of it in the larger array must be _larger_ than the end of the partition in the smaller array, _and vice versa_. 

If the partition is not correct, we move pointers. 

_NOTE_ solution from neetcode. 

```python
def findMedianSortedArrays(nums1: list[int], nums2: list[int]) -> float:
    # make sure that array a is smaller than be 
    a, b = nums1, nums2
    if (len(b)<len(a)):
        a,b=nums2,nums1
    
    total = len(a)+len(b)
    half = total // 2

    # run binary search on the smaller array
    l,r = 0,len(a)-1
    while True: # there is guaranteed to be a min.
        # compute middle value of array 'a' to get _left_ partition
        i = (l+r)//2 # for a
        j = half - i - 2 # get the middle of the 'b' array (substract 2 from len())
        # get values to check partitions
        aleft = a[i] if i >= 0 else float("-infinity")
        # if out of bound we need an entire array, so set to max
        aright = a[i+1] if i+1 < len(a) else float("infinity") # adjacent values
        bleft = b[j] if j >=0 else float("-infinity")
        # if out of bound, we need an entire array -> set infinity
        bright = b[j+1] if j+1 < len(b) else float("infinity")
        # note i,j can be OUT OF bounds
        if (aleft <= bright and bleft <= aright):
            # left partition is correct. Check cases
            if (total % 2):
                # compute median odd
                return min(aright, bright) # accounts for infinity
            else:
                # even size
                return (max(aleft, bleft) + min(aright, bright)) / 2
        elif (aleft > bright):
            # too many elements in 'a' -> reduce left partition
            r = i - 1
        else:
            l = i + 1



```
