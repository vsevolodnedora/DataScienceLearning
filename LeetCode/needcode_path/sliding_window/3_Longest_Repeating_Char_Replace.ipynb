# 424. Longest Repeating Character Replacement

You are given a string s and an integer k. You can choose any character of the string and change it to any other uppercase English character. You can perform this operation at most k times.

Return the length of the longest substring containing the same letter you can get after performing the above operations.

# Reasoning 

[neetcodevideo](https://www.youtube.com/watch?v=gqXU1UyA8pk)

This is from _Blind75_ list.  

Important problem.  

This problem can be solved in O(26 * n) time complexity which is O(n).

1. Use hashmap and count the number of occurances of each character in the __substring__, i.e., inside a _window_. 
2. Compute the len(substring) - count of most frequent element. It <= k, we can form a uniform string. If not, we cannot. In other words, we have a _sliding window_ and at each time we check if inside of this window, the len(window) - N of most common <= k. If True, we can form _a valid_ substring. 
3. Finding the _most frequent character_ requires looking throgh the `hash map`, which is O(26) time, as there are at maximum 26 entries for the enitre alphabet. __NOTE__ Even though we do it at each iteration, so we have O(26 * n) overall compolexity, __it is still a linear time__ solution. Just not super efficient. 
4. Overall algorithm: 
- Start with two pointers at i=0 and i=0; (incl. trivial case with opne cheracter)
- Shift right pointer while __condition__ for replacement is valid
- Shift left pointer while __condition__ is invalid
- Record the length of a subarray, i.e., distance between pointers at each iteration, update max.


```python
def characterReplacement(s: str, k: int) -> int:

    n = len(s)
    counts = {}
    l,r = 0, 1
    counts[s[0]] = 1
    res = 1
    while (r < n):
        # update map
        counts[s[r]] = counts.get(s[r], 0) + 1
        # check condition
        condition = (r - l + 1) - max(counts.values()) > k
        # move left if condition
        if condition:
            counts[s[l]] -= 1
            l+=1
        # compute result
        res = max(res,r-l+1)
        # move right pointer after
        r+=1
    return res



    n = len(s)
    counts = {}
    l,r = 0,1
    counts[s[0]] = 1
    res = 1
    while r < n:
        

        c_r = s[r]
        # add character to the hashmap
        if not (c_r in counts):
            counts[c_r] = 1
        else:
            counts[c_r] += 1

        # find the most freq. character
        c_main = None
        num = 0
        for (c, num_c) in counts.items():
            if (num_c > num): 
                    num = num_c
                    c_main = c
        # check the replacement condition
        tmp_ = (r - l + 1)
        tmp = tmp_ - num
        condition = tmp <= k
        if (condition):
            r+=1
        else:
            # remove cahracter from hashmap
            c_l = s[l]
            counts[c_l] -= 1
            l+=1
        
        # update result 
        res = max(res, r-l+1) # window size +1
        
    return res
print(characterReplacement(s = "A", k = 1), " Output: 1")
print(characterReplacement(s = "BB", k = 2), " Output: 2")
print(characterReplacement(s = "AA", k = 0), " Output: 2")
print(characterReplacement(s = "AAAA", k = 0), " Output: 4")
print(characterReplacement(s = "ABAB", k = 2), " Output: 4")
print(characterReplacement(s = "AABABBA", k = 1), " Output: 4")
```
